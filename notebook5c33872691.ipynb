```python
#step 1 import library
import pandas as pd
```

```python
#step 2 import data / read data
admission = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Admission%20Chance.csv')
```

```python
# describe target and features
admission.columns
```

```python
#setting target
y=admission['Chance of Admit ']

#setting features
X = admission[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']]
```

```python
print("Shape of X is : ",X.shape)
print("Shape of y is : ",y.shape)
```

```python
# step 4 : train test & split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2529 )
```

```python
# check shape of train and test sample
print("Shape of X_train is : ",X_train.shape)
print("Shape of X_test is : ",X_test.shape)
print("Shape of y_train is : ",y_train.shape)
print("Shape of y_test is : ",y_test.shape)
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
# calculating intercept and coefficient of model
print("Intercept is : ", model.intercept_)
print("Coefficient  is : ", model.coef_)
```

```python
#Step 7 : Predict model using predict() function
y_predict = model.predict(X_test)
```

```python
#Model Accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
print("Mean Absolute Error Of Model is : ",mean_absolute_error(y_test, y_predict))
```

```python
print("Mean Absolute Percentage Error Of Model is : ",mean_absolute_percentage_error(y_test, y_predict))
```

```python
print("Mean Square Error Of Model is : ",mean_squared_error(y_test, y_predict))
```
